#### üîß Goal:

**Scenario**: I already have a github repo set up from my macbook. How do I fetch it on my windows machine and make changes like adding a new file or funtion to existing code and push it back to github and pull the chnages in macboook?

#### üñ•Ô∏è 1. Clone the Repo on Windows

```bash
# Open terminal or Git Bash on Windows
git clone https://github.com/your-username/your-repo.git

# Navigate into the project folder
cd your-repo
```


#### ‚úçÔ∏è 2. Make Changes on Windows
Edit or add new files (e.g., update scripts, add notebooks, etc.)
```bash
# Stage changes
git add .

# Commit with a message
git commit -m "Describe your changes"

# Push changes to GitHub
git push origin main  # or 'master' depending on your branch
```

#### üçé 3. Pull Changes Back to MacBook
```bash
# On your MacBook terminal, navigate to the project directory
cd your-repo

# Pull the latest changes
git pull origin main  # or 'master'
```

To follow a very clean and professional Git workflow ‚Äî working on a feature in a new branch, then reviewing and merging it later.

**Scenario 2**: How to do everything on new branch called `win_codes` and then merge it after reviewing on my mackbook using github.

## üß† What We‚Äôll Do

We‚Äôll:

- ‚úÖ Clone the repo on Windows (if not already done)
- ‚úÖ Create a new branch `win_codes`
- ‚úÖ Add changes (file/function)
- ‚úÖ Commit and push branch to GitHub
- ‚úÖ On MacBook, pull the new branch and review
- ‚úÖ Merge `win_codes` ‚Üí `main` on GitHub
- ‚úÖ Pull merged changes to both machines

Let‚Äôs start step-by-step üëá

---

#### ü™ü On Your Windows Machine

#### ‚úÖ 1. Clone the repo (skip if already done)

```bash
git clone git@github.com:your-username/your-repo.git
cd your-repo
```
#### üåø 2. Create and switch to a new branch `win_codes`

```bash
git checkout -b win_codes
```
Now you're working on a new branch named `win_codes`.


#### ‚úèÔ∏è 3. Add a file or modify code
Example:
```bash
echo "print('This is a Windows feature')" > windows_script.py
```
üíæ 4. Stage and commit the changes
```bash
git add .
git commit -m "Added windows_script.py in win_codes branch"
```
üöÄ 5. Push this branch to GitHub
```bash
git push origin win_codes
```

    ‚úÖ Now the win_codes branch with your changes is visible on GitHub.
---

### üçé On Your MacBook
#### üîÉ 6. Fetch the new branch
```bash
cd path/to/your/local/repo
git fetch origin
```
#### üëÄ 7. Checkout the branch and review
``` bash
git checkout win_codes
```

Review the changes locally using:
```bash
git log --oneline
git diff main..win_codes # Dont do it on terminal instead on in Editor
```
Or open files in VS Code or your editor.

---
#### üîÄ 8. Merge using GitHub (best practice for code review)

1. Go to your GitHub repository in the browser.
2. You‚Äôll see a prompt like:

   > **"Compare & pull request"** for branch `win_codes`

3. Click **"Compare & pull request"**
4. Add a title and description for the pull request.
5. Click **"Create pull request"**
6. Review the code changes (diff)
7. Click **"Merge pull request"**
8. Confirm the merge ‚Üí ‚úÖ **Done**


### üîÑ 9. Pull Merged Changes Back to MacBook & Windows

**_üíª On MacBook:_**

```bash
git checkout main
git pull origin main
```
**_ü™ü On Windows:_**
git checkout main
git pull origin main
---
####üßπ (Optional) Clean Up the Feature Branch

**_Delete the local branch_**
git branch -d win_codes

**_Delete the remote branch_**
git push origin --delete win_codes

    ‚úÖ You're Done! Your code is now synced across both machines and the feature branch is cleaned up.
